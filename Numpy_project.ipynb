```python
import numpy as np

while True:
  print("\nMENU")
  print('*')
  print("1.ARITHMETIC OPERATIONS")
  print("2.STATISTICAL OPERATIONS")
  print("3.ARRAY MANIPULATION")
  print("4.EXIT")
  print('*')

  choice=int(input("Enter the choice:"))

  if choice==4:
    print("End")
    break

#-----BASIC ARITHMETIC OPERATIONS-----
  elif choice == 1:
    arr1=np.array(list(map(float,input("Enter Array1 Values:").split())))
    arr2=np.array(list(map(float,input("Enter Array2 Values:").split())))

    if arr1.shape!=arr2.shape:
      print("Arrays Must have the same shape")
      continue

    print("\nARITHMETIC RESULTS:")
    print("ADDITION:",arr1+arr2)
    print("SUBTRACTION:",arr1-arr2)
    print("MULTIPLICATION:",arr1*arr2)
    print("DIVISION:",np.divide(arr1,arr2,where=arr2!=0))

#-----STATISTICAL OPERATIONS

  elif choice==2:
    arr=np.array(list(map(float,input("Enter the array values:").split())))


    print("\nSTATISTICAL OPERATIONS RESULTS:")
    print("MEAN:",np.mean(arr))
    print("MEDIAN:",np.median(arr))
    print("STANDARD DEVIATION:",np.std(arr))
    print("VARIANCE:",np.var(arr))
    print("MAXIMUM:",np.max(arr))
    print("MINIMUM:",np.min(arr))
    print("SUM:",arr.sum())
    print("SHAPE:",arr.shape)
    print("SIZE:",arr.size)
    print("DATATYPE:",arr.dtype)

  #-----ARRAY MANIPULATION

  elif choice==3:
    arr=np.array(list(map(float,input("Enter the array values:").split())))

    if len(arr)==0:
      print("Array length cannot be zero")

    else:
      n=len(arr)
      factors=[i for i in range(1,n+1) if n%i==0]

      print("\nPOSSIBLE ARRAY COMBINATIONS ARE:")
      for r in factors:
        c=n//r
        print(f"({r}, {c})")

      rows,cols=map(int,input("Enter desired possible reshape options (rows,cols): ").split())

      if rows*cols!=n:
        print("Invalid! The number of elements in the array must be same")

      else:
        arr=arr.reshape(rows,cols)
        print("RESHAPED ARRAY:",arr)


      flat_arr=arr.flatten()
      print("SORTED ARRAY:",np.sort(flat_arr))
      print("REVERSE SORTED ARRAY:",np.sort(flat_arr)[::-1])
      print("UNIQUE VALUES IN THE ARRAY:",np.unique(flat_arr))
      print("FIRST ELEMENT IN THE ARRAY:",flat_arr[0])
      print("LAST ELEMENT IN THE ARRAY:",flat_arr[-1])
```
