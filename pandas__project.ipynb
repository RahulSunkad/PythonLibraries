PART-C:Data Cleaning

```python
#Task15

retail_data.dropna(subset=['CustomerID'], inplace=True)
```

```python
#Task16

retail_data['Description']=retail_data['Description'].fillna('Unknown Product')
```

```python
#Task18

retail_data.drop(retail_data[retail_data['Quantity'] <= 0].index, inplace=True)
```

```python
#Task19

retail_data.drop(retail_data[retail_data['UnitPrice'] <= 0].index,inplace=True)
```

```python
#Task20

retail_data.drop(
    retail_data[retail_data['InvoiceNo'].str.startswith('C')].index,
    inplace=True
)
```

```python
#Task21

retail_data.rename(columns={'Description':'ProductName'},inplace=True)
```

```python
#Task22
retail_data['InvoiceDate'] = pd.to_datetime(retail_data['InvoiceDate'])
```

```python
#Task23

retail_data['CustomerID']=retail_data['CustomerID'].astype(int)
```

```python
#Task24

retail_data.reset_index(drop=True,inplace=True)
```

```python
#Task25

retail_data['TotalPrice']=retail_data['UnitPrice']*retail_data['Quantity']
retail_data
```

```python
#Task31

retail_data['Quarter']='Q'+retail_data['InvoiceDate'].dt.quarter.astype(str)
```

```python
#Task32

retail_data['RevenueCategory']='Low'

retail_data.loc[(retail_data['TotalPrice']>=10) & (retail_data['TotalPrice']<=100),'RevenueCategory']='Medium'

retail_data.loc[(retail_data['TotalPrice']>100),'RevenueCategory']='High'
```

```python
#Task128

country_summary = retail_data.groupby('Country')['TotalPrice'].sum()

monthly_trends = retail_data.groupby(['Year', 'Month'])['TotalPrice'].sum()

top_50_customers = (
    retail_data.groupby('CustomerID')['TotalPrice']
    .sum()
    .sort_values(ascending=False)
    .head(50)
)
```
